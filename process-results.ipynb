```python
import pandas as pd
import os
import numpy as np
```

```python
def fixResults(resultsPath):
    overallResults = pd.read_csv(resultsPath + ".txt",sep="\t")
    badIndices = np.where(pd.isnull(overallResults['lower']))
    overallResults['lower'].loc[badIndices] = overallResults['upper'].loc[badIndices] - 0.1
    overallResults['Cat1_ID'] = overallResults['Cat1_ID'].astype(int)
    overallResults['Cat2_ID'] = overallResults['Cat1_ID'].astype(int)
    overallResults['Cat3_ID'] = overallResults['Cat1_ID'].astype(int)
    overallResults.to_csv(resultsPath + "-fixed.txt", index=None, sep="\t")
```

```python
fixResults("../../Results/rs2032598_C/results-combined")
```
